```python
import pandas as pd
import requests
import json
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
response = requests.get("http://api.worldbank.org/countries/indicators/SP.DYN.TFRT.IN?per_page=13900&date=1960:2016&format=json")
data = response.json()
data
```

```python
df = pd.read_json(json.dumps(data[1]))
df['country_cleaned'] = pd.DataFrame(map(lambda x: x['value'], list(df['country'])))
df['indicator_cleaned'] = pd.DataFrame(map(lambda x: x['value'], list(df['indicator'])))
df
```

```python
dfworld = df[df['country_cleaned'] == 'World']
plt.scatter(dfworld['date'], dfworld['value'])
```

```python
'''Againa, the data look very interesting. It initially increase, dips quite a lot, levels off very slightly,
the rate of decrease slows down, drops drastically and slows down by quite a lot.'''
```

```python
dfarab = df[df['country_cleaned'] == 'Arab World']
plt.scatter(dfarab['date'], dfarab['value'])
'''Here I can see how the fertility rate has changed in the Arab World.
This is quite different than for the whole world.
It was decreaseing very slightly and then started decreasing rapidly from 1980.
Then it levelled off around 2005.'''
```

```python
dfeurope = df[df['country_cleaned'] == 'Europe & Central Asia (developing only)']
plt.scatter(dfeurope['date'], dfeurope['value'])
'''This looks very interesting and different to the world data.
It decreases, then there is a sudden increase, steady decrease, levells off and decreases sharply.
In the last few years it increases a bit and then starts to decrease very slowly again.
This is very interesting and different and I am wondering why this is happening.'''
```

```python
dfssafrica = df[df['country_cleaned'] == 'Sub-Saharan Africa (all income levels)']
plt.scatter(dfssafrica['date'], dfssafrica['value'])
'''It is really interesting to see that the values here have been increasing
and then it starts to decrease from 1980 and has been decreasing ever since.'''
```

```python
dfceurope = df[df['country_cleaned'] == 'Central Europe and the Baltics']
plt.scatter(dfceurope['date'], dfceurope['value'])
'''This graph shows some really interesting variations.
However, it is worth noting that the variations are only between 2.6 and 1.4 child - so the
decrease in fertility rate is not actually as big in comparison to other regions.'''
```

```python
dfeasia = df[df['country_cleaned'] == 'East Asia & Pacific (all income levels)']
plt.scatter(dfeasia ['date'], dfeasia ['value'])
'''Here we can see a slight increase and thena  very sharp, rapid decrease, which levels off
and then decreases again. It then completely flattens out and is steady for a few years.'''
```
