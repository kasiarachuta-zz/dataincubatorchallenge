```python
import pandas as pd
import requests
import json
```

```python
'''Here I want to compare the data on female and male vulnerable employment.
According to the World Bank: Vulnerable employment is unpaid family workers and own-account 
workers as a percentage of total employment.
I expect this to be higher for women but I am not if it would increase/decrease or stay the same over time.'''

response = requests.get("http://api.worldbank.org/countries/indicators/SL.EMP.VULN.FE.ZS?per_page=13900&date=1960:2016&format=json")
data = response.json()

data
```

```python
# I now want to 

df = pd.read_json(json.dumps(data[1]))
df['country_cleaned'] = pd.DataFrame(map(lambda x: x['value'], list(df['country'])))
df['indicator_cleaned'] = pd.DataFrame(map(lambda x: x['value'], list(df['indicator'])))
df.head()
```

```python
# Now I just want to look at some particular  columns.

df = df[['date', 'value', 'country_cleaned']]
pd.isnull(df.value).value_counts()
```
