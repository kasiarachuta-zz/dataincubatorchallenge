```python
import pandas as pd
import requests
import json
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# I first import the data on females. I put in a DataFrame and clean it.

response = requests.get("http://api.worldbank.org/countries/indicators/SP.DYN.LE00.FE.IN?per_page=13900&date=1960:2016&format=json")
data = response.json()

df1 = pd.read_json(json.dumps(data[1]))
df1['country_cleaned'] = pd.DataFrame(map(lambda x: x['value'], list(df1['country'])))
df1['indicator_cleaned'] = pd.DataFrame(map(lambda x: x['value'], list(df1['indicator'])))

df1.head()
```

```python
# I then import the data on males.

response = requests.get("http://api.worldbank.org/countries/indicators/SP.DYN.LE00.MA.IN?per_page=13900&date=1960:2016&format=json")
data = response.json()

df2 = pd.read_json(json.dumps(data[1]))
df2['country_cleaned'] = pd.DataFrame(map(lambda x: x['value'], list(df2['country'])))
df2['indicator_cleaned'] = pd.DataFrame(map(lambda x: x['value'], list(df2['indicator'])))

df2.head()
```

```python
# I first want the data on the females in the world.

df1world = df1[df1['country_cleaned'] == 'World']

df1world.dropna()
```

```python
# I then want the data on the males in the world.

df2world = df2[df2['country_cleaned'] == 'World']

df2world.dropna()
```

```python
# I now want to plot the scatter plot on the data on females.

plt.scatter(df1world['date'], df1world['value'])
```

```python
# I now want to plot the scatter plot on the data on males.

plt.scatter(df2world['date'], df2world['value'])
# There is an interesting dip in 2003.
```

```python
# I add a column about gender and then concatenate the two tables.

df1world['Gender'] = 'Female'
df2world['Gender'] = 'Male'

frames = [df1world, df2world]

concatgender = pd.concat(frames)

concatgender.to_csv("concatgender.csv")

# This comes up with an error but actually works fine.
```
