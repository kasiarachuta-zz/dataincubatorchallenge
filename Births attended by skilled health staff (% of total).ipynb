```python
import pandas as pd
import requests
import json
```

```python
response = requests.get("http://api.worldbank.org/countries/indicators/SH.STA.BRTC.ZS?per_page=13900&date=1960:2016&format=json")
data = response.json()

data
```

```python
df = pd.read_json(json.dumps(data[1]))
df.head()
```

```python
# I want to clean the DataFrame

df['country_cleaned'] = pd.DataFrame(map(lambda x: x['value'], list(df['country'])))
df['indicator_cleaned'] = pd.DataFrame(map(lambda x: x['value'], list(df['indicator'])))
df
```
